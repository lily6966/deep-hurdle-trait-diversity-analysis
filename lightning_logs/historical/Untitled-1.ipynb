```python
import pandas as pd

# Load both CSV files
df1 = pd.read_csv('count_pred_results.csv')
df2 = pd.read_csv('presence_probability_results.csv')

# Keep the first two columns (e.g., location info)
location_cols = df1.iloc[:, :2]

# Select numeric columns to multiply (skip first two)
values1 = df1.iloc[:, 2:]
values2 = df2.iloc[:, 2:]

# Element-wise multiplication
result_values = values1 * values2
# Round to nearest integer
result_rounded = np.round(result_values).astype(int)

# Combine location columns with result
result_df = pd.concat([location_cols, result_rounded], axis=1)

# Save the result
result_df.to_csv('multiplied_result.csv', index=False)
result_df
```

```python
import numpy as np

# Assuming `data` is a NumPy array
print("Mean:", np.mean(result_df.iloc[:, 2:]))
print("Standard Deviation:", np.std(result_df.iloc[:, 2:]))
print("Min:", np.min(result_df.iloc[:, 2:]))
print("Max:", np.max(result_df.iloc[:, 2:]))
print("Median:", np.median(result_df.iloc[:, 2:]))
print("25th Percentile:", np.percentile(result_df.iloc[:,2:], 25))
print("75th Percentile:", np.percentile(result_df.iloc[:,2:], 75))
```

```python
# Load both CSV files
metrics = pd.read_csv('analysis_results/combined_metrics.csv')
metrics
```

```python
# Suppose the column is named 'F1 Score', and species names are in a column called 'Species'
# Step 1: Filter species with F1 Score > 0.3
filtered_species = metrics[metrics['F1 Score'] > 0.3]['Label']

# Step 2: Filter columns in multiplied that match those species (skip first two location columns)
location_cols = result_df.iloc[:, :2]
species_cols = result_df.loc[:, result_df.columns[2:]]  # species predictions

# Keep only columns for species in filtered list
filtered_species_cols = species_cols.loc[:, species_cols.columns.isin(filtered_species)]

# Combine with location info
filtered_result = pd.concat([location_cols, filtered_species_cols], axis=1)

# (Optional) Save or inspect
filtered_result.to_csv('filtered_multiplied_result.csv', index=False)
filtered_result.head()
```

```python
# Load both CSV files
df1 = pd.read_csv('count_pred_results.csv')
df2 = pd.read_csv('presence_probability_results.csv')
# Keep only columns for species in filtered list
filtered_species_count = df1.iloc[:, 2:].loc[:, species_cols.columns.isin(filtered_species)]
filtered_species_prob = df2.iloc[:, 2:].loc[:, species_cols.columns.isin(filtered_species)]

filtered_species_count = pd.concat([df1.iloc[:, :2], filtered_species_count], axis=1)
filtered_species_prob = pd.concat([df2.iloc[:, :2], filtered_species_prob], axis =1 )

# (Optional) Save or inspect
filtered_species_count.to_csv('filtered_count_result.csv', index=False)
filtered_species_prob.to_csv('filtered_presence_prob_result.csv', index=False)
filtered_species_count
```
